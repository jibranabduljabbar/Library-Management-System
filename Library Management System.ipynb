```python
class Library:
    
    def __init__(self, list_Of_Books):
        
        self.Available_Books = list_Of_Books
        
    def display(self):
        
        print("\n" + "Avaible Books In Library are: " + "\n")
    
        for book in self.Available_Books:
            print(book)
            
    def lend_Book(self, requested_Book):
        
        if requested_Book in self.Available_Books:
            print("\n" + f"You have now Borrowed the Book {requested_Book}" + "\n")
            self.Available_Books.remove(requested_Book)
        else:
            print("\n" + "We don't have that Book" + "\n")
            
    def add_Book(self, Book):
        
        if Book in self.Available_Books:
            print("\n" + "We have this Book Already" + "\n")
        else:
            print("\n" + "Thank you for the Book" + "\n")
            self.Available_Books.append(Book)
    
class Customer:
    
    def requested_Book(self):
        self.Book = input("\n" + "Enter the name of the Book you want: " + "\n")
        return self.Book
    
    def return_Book(self):
        self.Book = input("\n" + "Enter the name of the Book you want to return: " + "\n")
        return self.Book

library = Library(["Chemistry", "Physics", "Biology", "Mathematics", "Urdu", "English", "Pak Studies"])

customer = Customer()

while True:
    
    print("\n" + "Enter 1 to Display Books" + "\n")
    
    print("\n" + "Enter 2 to Lend a Book" + "\n")
    
    print("\n" + "Enter 3 to Return a Book" + "\n")
    
    print("\n" + "Enter 4 to Quit" + "\n")
    
    User = int(input(""))
    
    if User == 1:
        library.display()
        
    elif User == 2:
        requested_Book = customer.requested_Book()
        library.lend_Book(requested_Book)
        
    elif User == 3:
        return_Book = customer.return_Book()
        library.add_Book(return_Book)
        
    elif User == 4:
        break
```
